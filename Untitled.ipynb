```python
ela = pd.read_csv('data/aggregated_ela_representation_dropna.csv',index_col=[0,1,2])
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

from yellowbrick.cluster import KElbowVisualizer
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
ela_scaled=scaler.fit_transform(ela)
# Instantiate the clustering model and visualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,25))

visualizer.fit(ela_scaled)        # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=visualizer.elbow_value_)
```

```python
labels_kmeans = kmeans.fit_predict(ela_scaled)
```

```python
ela_scaled=pd.DataFrame(ela_scaled,index=ela.index)
```

```python
ela_scaled['kmeans_cluster']=labels_kmeans
```

```python
import numpy as np

from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering


def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)
```

```python
for metric in ['euclidean','cosine']:
    print(metric)
    agglomerative = AgglomerativeClustering(distance_threshold=0, n_clusters=None,affinity =metric
                                           ,linkage='ward' if metric=='euclidean' else 'average')

    agglomerative = agglomerative.fit(ela_scaled if metric=='euclidean' else ela)
    plt.title("Hierarchical Clustering Dendrogram")
    plot_dendrogram(agglomerative, truncate_mode="level", p=10)
    plt.xlabel("Number of points in node (or index of point if no parenthesis).")
    plt.show()
    
```

```python
for metric in ['euclidean','cosine']:
    print(metric)
    agglomerative = AgglomerativeClustering(distance_threshold=0, n_clusters=None,affinity =metric
                                           ,linkage='ward' if metric=='euclidean' else 'average')

    agglomerative = agglomerative.fit(ela_scaled if metric=='euclidean' else ela)
    ela_scaled[f'agglomerative_cluster_{metric}']=agglomerative.labels_
```

```python
import sklearn
cluster_labels=['kmeans_cluster','agglomerative_cluster_euclidean','agglomerative_cluster_cosine']
for cluster_label in cluster_labels:
    print(sklearn.metrics.silhouette_score(ela_scaled.drop(columns=cluster_labels, axis=1), ela_scaled[cluster_label]))
```
